```python
import pandas as pd
import seaborn as sns
```

```python
Master_File = pd.ExcelFile('Kinetic_Model_Fluxes.xlsx')
WT = pd.read_excel(Master_File, 'Base').T
WT['Type']='WT'
MT = pd.read_excel(Master_File, 'Third_order10').T
MT['Type']='MT'
```

```python
Combined=pd.concat([WT,MT],axis=0)
Combined=Combined.reset_index()
```

```python
Combined.drop(19,axis=0,inplace=True)
Combined.columns=Combined.iloc[0]
```

```python
Combined.drop(0,axis=0,inplace=True)
Combined
```

```python
sns.set_theme(style="whitegrid")

sns.violinplot('EX_co', hue="WT",data=Combined)
```
