```python
import pandas as pd
```

```python
WT=pd.read_excel('Kinetic_Model_Fluxes.xlsx',sheet_name='Base')
MT=pd.read_excel('Kinetic_Model_Fluxes.xlsx',sheet_name='Third_order10')
```

```python
WT
```

```python
WT['Median']=0
MT['Median']=0
```

```python
WT['Median']=WT.iloc[:,1:].median(axis=1)
MT['Median']=MT.iloc[:,1:].median(axis=1)
```

```python
WT=WT[['Row','Median']]

MT=MT[['Row','Median']]
```

```python
Compare_Table=pd.DataFrame(WT.index)
Compare_Table['Median_WT']=WT['Median']
Compare_Table['Median_MT']=MT['Median']
WT.set_index('Row',inplace=True)
MT.set_index('Row',inplace=True)
```

```python
Compare_Table.style.background_gradient(cmap='summer_r')
```

```python
WT.to_csv('Wild_Type.csv')
MT.to_csv('Mutant.csv')
```
